```python
import os
from pathlib import Path
from ultralytics import YOLO

# Define the paths
model_path = "/content/drive/MyDrive/myntra-yolov8/runs/detect/train/weights/best.pt"
input_dir = Path("/content/drive/MyDrive/myntra-yolov8/data/test/images")
output_dir = Path("/content/drive/MyDrive/myntra-yolov8/Results")

# Create the output directory if it doesn't exist
output_dir.mkdir(parents=True, exist_ok=True)

# Load the YOLO model
model = YOLO(model_path)

# Get list of all image files in the input directory
image_paths = list(input_dir.glob('*.jpg')) + list(input_dir.glob('*.jpeg')) + list(input_dir.glob('*.png'))

# Run inference on each image and save the results
for img_path in image_paths:
    results = model([str(img_path)])  # Run inference on the image

    for result in results:
        # Construct the output file path
        output_file_path = output_dir / img_path.name

        # Save the result image to the output directory
        result.save(filename=str(output_file_path))  # Save the image to the specified output directory
```
